```python
import numpy as np

# Function to fill NaN values with random values from the same column
def fill_nan_with_random(column):
    mask = column.isnull()
    num_nulls = mask.sum()
    random_values = np.random.choice(column[~mask], num_nulls)
    column[mask] = random_values
    return column

# Apply the function to the DataFrame
data['FIQ'] = fill_nan_with_random(df['FIQ'])

print(data)





```
